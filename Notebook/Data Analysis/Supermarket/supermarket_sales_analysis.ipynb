```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sales = pd.read_csv('supermarket_sales.csv')
```

```python
sales.head(10)
```

```python
sales.columns = sales.columns.str.replace(' ','_')
sales
```

```python
sales['Product_line']
```

## 1. which city has more frmales shopping?

```python
sales.groupby(['City', 'Gender']).count()
```

```python
female_shoppers = sales.groupby(['Gender', 'City']).count()['Customer_type']
female_shoppers
```

```python
female_shoppers.unstack(level = 0).plot(kind = 'bar')
plt.legend(loc = 'lower right')
```

## 2. who spend more, men or women?

```python
sales.groupby('Gender').sum()
```

```python
spend_gender = sales.groupby('Gender').sum()['Total']
spend_gender
```

```python
spend_gender.plot(kind = 'bar')
```

## 3. which customer type spend most, member or normel?

```python
sales.groupby('Customer_type').sum()
```

```python
spend_customer_type = sales.groupby('Customer_type').sum()['Total']
spend_customer_type
```

```python
spend_customer_type.plot(kind = 'bar')
```

## 4. which product line sells more

```python
sales.groupby('Product_line').count()['Invoice_ID']
```

```python
product_line_sells = sales.groupby('Product_line').count()['Invoice_ID']
product_line_sells.plot(kind = 'bar')
plt.grid()
```

## 5. which product line is popular among men vs women

```python
sales.groupby(['Gender', 'Product_line']).count()
```

```python
gender_product_line = sales.groupby(['Gender', 'Product_line']).count()['Invoice_ID']
gender_product_line
```

```python
gender_product_line.unstack(level = 0).plot(kind = 'bar')
plt.legend(loc = 'lower right')
plt.grid(axis = 'y')
```

## 6. what days of the month make most sales
## 7. what month make most sales

```python
sales['Date']
```

```python
pd.to_datetime(sales['Date'])
```

```python
pd.to_datetime(sales['Date']).dt.day
pd.to_datetime(sales['Date']).dt.month

sales['Day'] = pd.to_datetime(sales['Date']).dt.day
sales['Month'] = pd.to_datetime(sales['Date']).dt.month

sales.head(10)
```

```python
#days
sales.groupby('Day').sum()['Total']
```

```python
days_sales = sales.groupby('Day').sum()['Total']
days_sales.plot()
plt.grid()
```

```python
#month
sales.groupby('Month').sum()['Total']
```

```python
months_sales = sales.groupby('Month').sum()['Total']
months_sales.plot(kind = 'bar')
plt.grid(axis = 'y')
```

## 8. what hour make most sales

```python
sales['Time']
```

```python
pd.to_datetime(sales['Time'])
```

```python
pd.to_datetime(sales['Time']).dt.hour

sales['Hour'] = pd.to_datetime(sales['Time']).dt.hour

sales.head(10)
```

```python
#hours
hours_sales = sales.groupby('Hours').sum()['Total']
hours_sales.plot()
plt.grid()
```

## 9. what time do people make more E-payment vs what time do people make cash payment

```python
sales.groupby('Payment').count()
```

```python
sales.groupby(['Payment', 'Hour']).count()['Customer_type']
```

```python
sales.groupby(['Payment', 'Hour']).count()['Customer_type'].unstack(level = 0).plot(kind = 'bar')
plt.grid(axis = 'y')
plt.legend(loc = 'lower right')
```
